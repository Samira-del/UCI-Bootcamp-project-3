```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
import requests
from bs4 import BeautifulSoup as bs
import pprint as p
from numpy import array
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
```

# Read CSV file

```python
csv_file = "../data/true_car_listings.csv"
car_list_df = pd.read_csv(csv_file)
car_list_df.head()
```

```python
data_type = car_list_df.dtypes
print(data_type)
```

# Data Preparnig

```python
a = car_list_df.groupby (['Make']).size()
b = car_list_df.groupby ('Make')['Price'].max()
c = {'count': a,
     'average': b}
df = pd.DataFrame(c)
df.sort_values(by=['count'], inplace=True)
df
```

```python
car_list_df = car_list_df.drop(car_list_df[car_list_df.Price > 80000].index)
# car_list_df = car_list_df[car_list_df.Make == 'Nissan']
```

```python
Age = 2020 - car_list_df['Year']
car_list_df['Age'] = Age
car_list_df.head()
```

```python
y = car_list_df['Price']
plt.figure(figsize=(8,4))
sns.distplot(y).set_title('Frequency Distribution Plot of Prices')
```

```python
car_list_df_drop = car_list_df.drop(columns=['Price', 'Vin', 'City'], axis=1)
sns.boxplot(y=y, data=car_list_df_drop, width=0.5)
```

```python
values = array(car_list_df_drop['Make'])
label_encoder = LabelEncoder()
car_list_df_drop['Make'] = label_encoder.fit_transform(values)

values = array(car_list_df_drop['Model'])
label_encoder = LabelEncoder()
car_list_df_drop['Model'] = label_encoder.fit_transform(values)

values = array(car_list_df_drop['State'])
label_encoder = LabelEncoder()
car_list_df_drop['State'] = label_encoder.fit_transform(values)

car_list_df_drop
```

# Define Target and Features

```python
y = car_list_df['Price']
x = car_list_df_drop
x.head()
```

```python
#The mean and median number is very differet meaning that we have outliner or extrem data.
print(y.mean())
print(y.median())
```

```python
print(y.max())
print(y.min())
```

# Define Train and Test Data

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
X_train.head()
```

# make Multilinear Regression Model

```python
n_features = 6
X, y = make_regression(n_samples=len(X_train), n_features=n_features, 
                       n_informative=n_features, random_state=42, 
                       noise=0.5, bias=100.0)
print(X.shape)

model = LinearRegression()

# Fitting our model with all of our features in X
model.fit(X, y)

score = model.score(X, y)
print(f"R2 Score: {score}")
```

# Try 3 types of Normalize

```python
X_scaler  = MinMaxScaler().fit(X_train)
X_train_scaled  = X_scaler.transform(X_train)
X_test_scaled  = X_scaler.transform(X_test)

model.fit(X_train_scaled, y_train)
training_score = model.score(X_train_scaled, y_train)
testing_score = model.score(X_test_scaled, y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```
